```python
'''
Find the factors of an integer
'''

def factor(b):
    for i in range(1, b+1):
        if b % i == 0:
            print(i)
            
if __name__ == '__main__':
    b = input('Your number please: ')
    b = float(b)
    if b>0 and b.is_integer():
        factor(int(b))
    else: 
        print('Please enter a positive integer')

```

```python
'''
Multiplication table printer
'''

def multi_table(a):
    for i in range(1, 11):
        print('{0} × {1} = {2}'.format(a, i, a*i))

if __name__ == '__main__':
    a = input('Enter a number: ')
    multi_table(float(a))
```

```python
'''
Converting units of measurements
Unit converter: Miles and Kilometers
'''

def print_menu():
    print('1. Kilometers to Miles')
    print('2. Miles to Kilometres')

def km_miles():
    km = float(input('Distance in kilometers: '))
    miles = km / 1.609
    
    print('Distance in miles: {0}'.format(miles))
    
def miles_km():
    miles = float(input('Distance in Miles: '))
    km = miles * 1.609
    
    print('Distance in kilometres: {0}'.format(km))
    
    
if __name__ == '__main__':
    print_menu()
    choice = input('Which conversion would you like to do?: ')
    if choice == '1':
        km_miles()
    if choice == '2':
        miles_km()    
        
    
```

```python
'''
Quadratic equation root calculator
'''
import math


def roots(a, b, c):
    D = (b**2 - 4 * a * c)**0.5
    x_1 = (-b + D) / 2*a
    x_2 = (-b - D) / 2*a
    
    print('x1: {0}'.format(x_1))
    print('x2: {0}'.format(x_2))

if __name__ == '__main__':
    a = input('a: ')
    b = input('b: ')
    c = input('c: ')
    roots(float(a), float(b), float(c))
    
```

```python
'''
Even-odd vending machine
Get a input and print wheter the number is even or odd. Print the next 9 even/odd numbers
'''

def even_odd(a):
    if a%2 ==0:
        print('Number is even')
    else:
        print('Number is odd')
        
def next_numbers(a):
    for i in range(a, a+20, 2):
        print(i)
        
if __name__ == '__main__':
    a = input('Enter a number: ')
    b = int(a)
    even_odd(b)
    next_numbers(b)
```

```python
'''
Enhanced multiplication table printer
'''

def multi_table(a, b):
    for i in range(1, b+1):
        print('{0} × {1} = {2}'.format(a, i, a*i))
        
if __name__ == '__main__':
    a = input('Enter a number: ')
    b = input('Enter the amount of multiples: ')
    multi_table(int(a), int(b))
```

```python
'''
Fraction calculator

'''


from fractions import Fraction as fr


def add(a, b):
    print('Result of addition is: {0}'.format(a+b))
    
def subtract(a, b):
    print('Result of substractions is: {0}'.format(a-b))
    
def divide(a, b):
    print('Result of division is: {0}'.format(a/b))
    
def multiply(a, b):
    print('Result of multiplication is: {0}'.format(a*b))
    
    
if __name__ == '__main__':
    a = fr(input('Enter the first fraction: '))
    b = fr(input('Enter the second friction: '))
    op = input('Operation to perform - Add, Subtract, Divide, Multiply: ')
    if op=='Add':
        add(a, b)
    if op=='Subtract':
        subtract(a, b)
    if op=='Divide':
        divide(a, b)
    if op=='Multiply':
        multiply(a, b)
```

```python
'''
Multiplication table printer with
exit power to user
'''

def multi_table(a, b):
    for i in range(1, b+1):
        print('{0} × {1} = {2}'.format(a, i, a*i))
        
if __name__ == '__main__':
    while True:
        a = input('Enter a number: ')
        b = input('Enter the amount of multiples: ')
        multi_table(int(a), int(b))
        
        answer = input('Do you want to exit? (y) for yes ')
        if answer == 'y':
            break
            
```
